```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.random.seed(2343243)
```

```python
mean_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1 = np.random.multivariate_normal(mean_vec1, cov_mat1, 100)
```

```python
mean_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0], [0,1,0],[0,0,1]])
class2 = np.random.multivariate_normal(mean_vec2, cov_mat2, 100)
```

```python
from mpl_toolkits.mplot3d import Axes3D, proj3d
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot(class1[:,0], class1[:,1], class1[:,2],'o')
ax.plot(class2[:,0],class2[:,1],class2[:,2],'^')
plt.show()
```

```python
all_data = np.concatenate((class1, class2))
```

```python
pca = PCA()
transformed_data = pca.fit_transform(all_data)
transformed_data
```

```python
pca1 = PCA(n_components = 2)
tranformed_data = pca1.fit(all_data)
transformed_data
```

```python
pca1.components_
```

```python
plt.plot(transformed_data[:100,0], transformed_data[:100,1],'^')
plt.plot(transformed_data[100:,0], transformed_data[100:,1],'o')
plt.show()
```

```python
x_approx = pca.inverse_transform(transformed_data)
x_approx
```

## PCA FROM SCRATCH

```python
all_data_t = all_data.T
```

```python
con = np.cov(all_data_t)
```

```python
con
```

```python
con.shape
```

```python
eigen_val, eigen_vector = np.linalg.eig(con)
```

```python
eigen_val_vector_pair = []
for i in range(len(eigen_val)):
    eigen_vec = eigen_vector[:,1]
    eigen_val_vector_pair.append((eigen_val[i], eigen_vec))
```

```python
eigen_val_vector_pair.sort(reverse=True)
eigen_val_vector_pair
```
