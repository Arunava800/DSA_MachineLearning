```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data/data.csv', delimiter = ',')
```

```python
data
```

```python
x = data[:,0]
```

```python
y = data[:,1]
```

```python
x_reshape = x.reshape(100,1)
y_reshape = y.reshape(100,1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_reshape, y_reshape, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
coefficient = model.coef_
```

```python
intercept = model.intercept_
intercept
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols = 1, figsize=(10,10))
scatter = ax1.scatter(x= x_train,
            y=y_train,
            c=y_train)
line = coefficient * x_train + intercept
ax1.plot(x_train, line,'g')
ax1.set(title="X_train vs Y_train", xlabel = "x_train", ylabel="y_train")
ax1.grid(True)
fig.add_gridspec(2,1,hspace=3)
scatter = ax2.scatter(x=x_test,
                     y=y_test,
                     c=y_test)
line1 = coefficient * x_test + intercept
ax2.plot(x_test,line1,'b')
ax2.set(title='X_test vs. Y_test', xlabel='x_test', ylabel='y_test')
ax2.grid(True)
```

```python
# Scoring the data
score_test = model.score(x_test,y_test)
score_train = model.score(x_train, y_train)
score_test, score_train
```
