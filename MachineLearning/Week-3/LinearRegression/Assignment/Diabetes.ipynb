```python
x = train_data.drop('Y', axis = 1)
```

```python
y = train_data.Y
```

```python
x.shape, test_data.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(x,y, test_size = 0.33, random_state = 42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
alg = LinearRegression()
alg.fit(x_train, y_train)
```

```python
train_score = alg.score(x_train, y_train)
test_score = alg.score(x_test, y_test)
print(f"Training Score {train_score:.5f}")
print(f"Test Score {test_score:.5f}")
```

```python
slope = alg.coef_
intercpet = alg.intercept_
print(f"The coefficient is {slope.mean():.5f}")
print(f"The intercept is {intercept:.5f}")
```

```python
y_preds = alg.predict(test_data)
predicted_score = alg.score(test_data, y_preds)
print(f"The predicted score is {predicted_score*100:.5f}")
```
