# Missing Values with Scikit-Learn

```python
car_missing_extended = pd.read_csv("data/car-sales-missing-data.csv")
```

```python
car_missing_extended
```

```python
car_missing_extended.dropna(subset=["Price"], inplace=True)
```

```python
car_missing_extended
```

```python
car_missing_extended.isna().sum()
```

```python
x = car_missing_extended.drop("Price", axis = 1)
```

```python
y = car_missing_extended["Price"]
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
cat_imputer = SimpleImputer(strategy="constant", fill_value="missing")
door_imputer = SimpleImputer(strategy = "constant", fill_value = 4)
numerical_imputer = SimpleImputer(strategy = "mean")
```

```python
cat_features = ['Make', 'Colour']
door_features = ['Doors']
num_features = ['Odometer']
```

```python
imputer = ColumnTransformer([
    ("cat_imputer", cat_imputer, cat_features),
    ("door_impute",door_imputer, door_features),
    ("num_imputer", numerical_imputer, num_features)
])
```

```python
filled_x = imputer.fit_transform(x)
```

```python
filled_x
```

```python
car_sales_filled = pd.DataFrame(filled_x)
```

```python
car_sales_filled
```

```python
car_sales_filled.columns = ["Make", "Colour","Door", "Odometer"]
```

```python
car_sales_filled
```
