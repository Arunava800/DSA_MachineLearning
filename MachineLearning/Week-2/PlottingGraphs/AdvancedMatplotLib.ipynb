##### Making figure with numpy arrays

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(x,np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond Butter":10,
                    "Peanut Butter":8,
                    "Cashew Butter": 12}
fig, ax = plt.subplots(figsize=(10,10))
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Nut Butter Prices", xlabel = "Products", ylabel= "Prices");
```

#### Horiziontal Bar Graph, Histogtams and Subplots

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),(nut_butter_prices.values()));
```

```python
# Make some data
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

#### Two options for subplots

```python
fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize=(10,5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Second option
fig, ax = plt.subplots(nrows=2, ncols = 2, figsize=(10,5))
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000))
```

## Plotting from Pandas DataFrame

```python
# Make a data frame
car_sales = pd.read_csv("car-sales.csv")
```

```python
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]','', regex=True)
```

```python
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].astype(float)
```

```python
car_sales 
```

```python
car_sales["Sales_Date"] = pd.date_range('1/1/2020', periods = len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total_Sales"] = car_sales.Price.cumsum()
```

```python
car_sales
```

```python
# Plot the data
car_sales.plot(x="Sales_Date", y="Total_Sales");
```

```python
car_sales.plot(x="Odometer (KM)", y="Price",kind="scatter")
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist();
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,50),subplots=True);
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind="scatter", x='age', y='chol', c='target');
```

```python
fig, ax = plt.subplots(figsize=(10,5))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

# ax.set_xlim([45,100])
```

## OO method

```python
fig, ax = plt.subplots(figsize=(10,5))
scatter = ax.scatter(x=over_50['age'],
                    y = over_50['chol'],
                    c=over_50['target']);

ax.set(title="Heart Disease vs Cholesterol Levels",
      xlabel = "Age",
      ylabel= "Cholesterol");
#add a legend
ax.legend(*scatter.legend_elements(),title="Target")
# add a horiziontal line
ax.axhline(over_50["chol"].mean(),
          linestyle = '-.');

```

```python
fig, (ax0, ax1) = plt.subplots(nrows= 2, ncols= 1,
                              figsize= (10,10),
                              sharex=True)

scatter = ax0.scatter(x=over_50['age'],
                     y = over_50['chol'],
                     c = over_50['target'])
ax0.set(title="Heart Disease vs Cholesterol", ylabel= "Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(over_50["chol"].mean(),
          linestyle="-.")

scatter1 = ax1.scatter(x=over_50['age'],
                     y = over_50['thalach'],
                     c = over_50['target'])
ax1.set(title="Heart Disease vs Thalch", xlabel = "age", ylabel="Thalch")
ax1.legend(*scatter1.legend_elements(), title="Target")
ax1.axhline(over_50["thal"].mean(),
           linestyle="-.");

#Add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight= "bold");
```
