## Customizing Graphs

```python
x = np.arange(0, 5, 0.1)
y = x ** 3
plt.plot(x,y, color="black", marker="p", linewidth = 0.5, label="X^3")
y = x ** 2
plt.plot(x,y, color="blue", label= "X^2")
plt.ylabel("y")
plt.xlabel ("X")
#plt.axis([0, 10, 0, 150])
plt.title("Matplotlib Demo")
plt.grid()
plt.text(2,8,"Marked",fontsize=12)
plt.legend()
plt.show()
```

## Pie Graphs

```python
sizes = [3,4,6,2]
color = ["blue","purple","pink","green"]
labels = ["Class A", "Class B", "Class C", "Class D"]
plt.title("Split among titles")
ex = [0, 0, 0.1, 0]
plt.pie(sizes, colors = color, labels = labels, autopct = "%.1f%%", explode = ex,counterclock=False, startangle=90)
plt.axis("equal")
plt.show()
```
