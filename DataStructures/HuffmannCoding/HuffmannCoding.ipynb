```python
import heapq
import os
```

```python
class BinaryTreeNode:
    def __init__(self, value: str, frequency: int) -> None:
        self.value = value
        self.frequency = frequency
        self.left = None
        self.right = None

    def __lt__(self, other: object) -> bool:
        return self.frequency < other.frequency

    def __eq__(self, other: object) -> bool:
        return self.frequency == other.frequency
```
